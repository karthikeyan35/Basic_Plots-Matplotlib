# Bar Graph

A barplot (or barchart) is one of the most common type of plot. It shows the relationship between a numerical variable and a categorical variable. For example, you can display the height of several individuals using bar chart. Barcharts are often confounded with

The bars are positioned at x with the given alignment. Their dimensions are given by width and height. The vertical baseline is bottom (default 0).

Each of x, height, width, and bottom may either be a scalar applying to all bars, or it may be a sequence of length N providing a separate value for each bar

```python
#matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center')
x1=[1,2,3,4]
y1=[5,2,7,8]
x2=[1,2,3,4]
y2=[1,2,3,4]

plt.bar(x1,y1,label='Blue')
plt.bar(x2,y2,bottom=y1,label='Orange')
plt.title('Runs vs overs')
plt.xlabel('overs')
plt.ylabel('runs')
plt.legend()
plt.show()
```

```python
kar=DataFrame({'Revenue':[200,300,350,400,120,600]},index=['wipro','cisco','google','yahoo','facebook','tcs'])
```

```python
x=kar.index #taking data from data frame
y=kar['Revenue']
plt.bar(x,y)
plt.xticks(rotation=45)
plt.yticks(rotation=45)#Rotation of ticks
plt.title('company vs revenue')
plt.show()
```

```python
x=kar.index
y=kar['Revenue']
plt.barh(x,y)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.title('company vs revenue')
plt.show()
```

```python
#comparing two quantities 
india = (20, 35, 30, 35, 27)
eng = (25, 32, 34, 20, 25)
ind = np.arange(len(india))  # the x locations for the groups
width = 0.35  # the width of the bars
plt.bar(ind - width/2, india, width,label='India')
plt.bar(ind + width/2, eng, width,label='England')
plt.ylabel('Runs')
plt.title('Runs per Over')

plt.xticks(ind,('O1', 'O2', 'O3', 'O4', 'O5'))
plt.legend()
plt.show()
```

```python
#stackedbargraph
N = 5
men = (20, 35, 30, 35, 27)
women = (25, 32, 34, 20, 25)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

plt.bar(ind, men, width,bottom=5,label='men')
plt.bar(ind, women, width,bottom=men,label='women')#bottom gives the position from xaxis

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 5))
plt.legend()

plt.show()
```

```python
#horizontal bar graph
men = (20, 35, 30, 35, 27)
women = (25, 32, 34, 20, 25)

ind = np.arange(len(men))  # the x locations for the groups
width = 0.35  # the width of the bars
plt.barh(ind - width/2, men, width,label='Men',color='green')
plt.barh(ind + width/2, women, width,label='Women')
plt.xlabel('Scores')
plt.title('Scores by group and gender')
plt.yticks(ind,('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
plt.show()
```

# Box Plot

Make a box and whisker plot for each column of x or each vector in sequence x. The box extends from the lower to upper quartile values of the data, with a line at the median. The whiskers extend from the box to show the range of the data. Flier points are those past the end of the whiskers.

matplotlib.pyplot.boxplot(x, notch=None)

```python
np.random.seed(196)
x=np.random.rand(50) * 100
plt.title('Basic box  Plot')
plt.boxplot(x)
plt.show()
```

```python
x=df1['size']
plt.xlabel("Size")
x1=df1['size'].quantile(0.75)
y1=df1['size'].quantile(0.25)
ma=((x1-y1)*1.5)+x1
print(ma)


plt.title('Basic box  Plot')
red_square = dict(markerfacecolor='r', marker='s')
plt.boxplot(x,flierprops=red_square, vert=False)
plt.show()
```

```python
x=df1['tip']
y=df1['size']
z=[x,y]
plt.title('Basic box  Plot')
red_square = dict(markerfacecolor='r', marker='D')
plt.boxplot(z,flierprops=red_square,notch=True,showfliers=False)
plt.show()
print(df1['tip'].max())
```

# Pie Plot



Pie chart  is a circular graphic which is divided into slices to illustrate numerical proportion. The point of a pie chart is to show the relationship of parts out of a whole. 

The wedges are plotted counterclockwise, by default starting from the x-axis.

matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False,
labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, 
textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, data=None)

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
#plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df1['day'].unique()
sizes = df1['day'].value_counts()
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Sun')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
